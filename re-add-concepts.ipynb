```python
import json
import requests
from concurrent.futures import ThreadPoolExecutor, as_completed
```

```python
with open("all_uvl-procedures_references.json") as f:
    procedures = json.load(f)

with open("all_uvl-drug-entry_references.json") as f:
    drug_entry = json.load(f)

with open("all_uvl-dx_references.json") as f:
    dx = json.load(f)

with open("all_uvl-labtests_references.json") as f:
    labtests = json.load(f) 

with open("all_uvl-forms_references.json") as f:
    forms = json.load(f)

with open("all_uvl-allergens_references.json") as f:
    allergens = json.load(f)

with open("all_uvl-vital-signs_references.json") as f:
    vital_signs = json.load(f)

with open("all_uvl-cause-of-death_references.json") as f:
    cause_of_death = json.load(f)

with open("all_uvl-immunizations_references.json") as f:
    immunizations = json.load(f)

with open("all_uvl-registration_references.json") as f:
    registration = json.load(f)

with open("all_uvl-drugs_references.json") as f:
    drugs = json.load(f)
```

```python
def filter_concepts(references):
    return list(filter(lambda reference: reference["reference_type"] == "concepts", references))

def map_to_concept_url(references):
    return list(map(lambda reference: reference["expression"], references))

procedure_concepts = map_to_concept_url(filter_concepts(procedures['all_references']))
drug_entry_concepts = map_to_concept_url(filter_concepts(drug_entry['all_references']))
dx_concepts = map_to_concept_url(filter_concepts(dx['all_references']))
labtests_concepts = map_to_concept_url(filter_concepts(labtests['all_references']))
forms_concepts = map_to_concept_url(filter_concepts(forms['all_references']))
allergens_concepts = map_to_concept_url(filter_concepts(allergens['all_references']))
vital_signs_concepts = map_to_concept_url(filter_concepts(vital_signs['all_references']))
cause_of_death_concepts = map_to_concept_url(filter_concepts(cause_of_death['all_references']))
immunizations_concepts = map_to_concept_url(filter_concepts(immunizations['all_references']))
registration_concepts = map_to_concept_url(filter_concepts(registration['all_references']))
drugs_concepts = map_to_concept_url(filter_concepts(drugs['all_references']))
```

```python
procedure_references_urls = procedures["references_urls"]
drug_entry_references_urls = drug_entry["references_urls"]
dx_references_urls = dx["references_urls"]
labtests_references_urls = labtests["references_urls"]
forms_references_urls = forms["references_urls"]
allergens_references_urls = allergens["references_urls"]
vital_signs_references_urls = vital_signs["references_urls"]
cause_of_death_references_urls = cause_of_death["references_urls"]
immunizations_references_urls = immunizations["references_urls"]
registration_references_urls = registration["references_urls"]
drugs_references_urls = drugs["references_urls"]

all_references = list(set([
  *drugs_references_urls
]))

with open('all_uvl-references.json', 'w') as f:
    json.dump(all_references, f, indent=4)



```
